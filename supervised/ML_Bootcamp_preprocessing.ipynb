```python
#Import bibliotek
import pandas as pd
import numpy as np
import sklearn
```

```python
#Wygenerowanie danych
data = {
    'size':['XL','L','M','L','M'],
    'color':['red','green','blue','green','red'],
    'gender':['female','male','male','female','female'],
    'price':[199.0, 89.0, 99.0, 129.0, 79.0],
    'weight':[500, 450, 300, 380, 410],
    'bought':['yes','no','yes','no','yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
#Utworzenie kopii danych
df = df_raw.copy()
df.info()
```

```python
#zmiana typu danych i wstępna eksploracja
for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')
df.info()
```

```python
#podstawowe statystyki na temat data frame - domyślnie metoda wyświetla statystyki dla kolumn numerycznych
df.describe()
```

```python
#.T - transpozycja czyli zmiana układu data frame
df.describe().T
```

```python
#wyświetlanie statystyk o danych nienumerycznych
df.describe(include=['category']).T
```

#Label Encoder - encoding zmiennej docelowej na dane numeryczne

```python
from sklearn.preprocessing import LabelEncoder

#Label Encoder używamy do przekształcania tylko zmiennych docelowych które chcemy przewidzieć!!!
#w kolumnie 'bought' zmieniamy 'yes' na 1 oraz 'no' na 0
le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
#skrócona wersja Label Encoding
le.fit_transform(df['bought'])
```

```python
#w przypadku większej ilości klas możemy wyświetlić jak je zmapowano -> indeksy w macierzy odpowiadaja liczbie?
le.classes_
```

```python
#przypisanie zmiany do data frame
df['bought']=le.fit_transform(df['bought'])
df
```

```python
#możemy przywrócić dane do pierwotnej formy (sprzed Lable Encoding)
# df['bought'] = le.inverse_transform(df['bought'])
```

#One Hote Encoder - encoding pozostałych zmiennych na dane numeryczne

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop = 'first', sparse=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

#Pandas get_dummies() - lepsza metoda niż OneHot Encoding

```python
df
```

```python
#dzięki Label Encoder kolumna 'bought' nie podlega już komendzie gdyż jest numeryczna, bez tego też byłaby rodzielona na 2
pd.get_dummies(data=df)
```

```python
#drop_first=True, skraca nam liczbę kolumn jak poniżej
pd.get_dummies(data=df,drop_first=True)
```

```python
#prefix zmienia nam nazwe kolumny tak jak chcemy
#prefix_sep - zamienia podkreślnik na to co chcemy
pd.get_dummies(data=df, drop_first = True, prefix='new', prefix_sep='-')
```

```python
#get_dummies dla wybranej kolumny
pd.get_dummies(data=df, drop_first=True, columns=['size'])
```

#Standaryzacja - StandardScaler

```python
#standaryzacja
from sklearn.preprocessing import scale
scale(df['price'])
```

```python
#Standard Scaler jest lepszy od Scale, gdyż pozwala zachować na przyszłość wartości statystyk, można to wykorzystać w nowych danych
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```

```python
#standaryzacja kilku kolumn jednoscześnie
scaler = StandardScaler()
df[['price','weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
